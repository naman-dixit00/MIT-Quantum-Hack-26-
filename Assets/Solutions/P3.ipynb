```python
!pip install qiskit qiskit-aer
import qiskit
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

# 1. Load the circuit
qc = QuantumCircuit.from_qasm_file('P3_tiny_ripple.qasm')

# 2. Setup the Simulator for SPEED
# We use 'matrix_product_state' but limit the bond dimension to speed up calculations
sim = AerSimulator(
    method='matrix_product_state',
    matrix_product_state_max_bond_dimension=16  # Aggressive compression for speed
)

# 3. Optimize the circuit BEFORE running
# optimization_level=3 is the highest; it simplifies gates to make them run faster
qc_fast = transpile(qc, sim, optimization_level=3)
qc_fast.measure_all()

# 4. Use very low shots
# Because the "Peak" is a mathematical absolute in these puzzles,
# a few shots are usually enough to hit the global maximum.
result = sim.run(qc_fast, shots=10).result()
counts = result.get_counts()

# 5. Extract the winner
peak_bitstring = max(counts, key=counts.get)
print(f"Peak Bitstring: {peak_bitstring}")
```
